## Data types and missing values

#### Dtypes
this is the data type for a column in a DataFrame or a series

```python
# you can use the dtype property to grab the type of specific column:
reviews.price.dtype #this gets the dtype of the price column
```

note that columns consisting only of strings do not get their own type, they're given the _object_ type  
its possible to convert a column of one type into another wherever you need to using the **astype()** function

```python
#we can tansform the points column from its existing int64 to a float64 data type
reviews.points.astype('float64')

#a DataFrame has its own dtype too:
reviews.index.dtype
```

## Missing data
- entries missing values are given the value _NaN_, short for "Not a Number". These NaN values are always of the float64 dtype  
- to select NaN entries you can use **pd.isnull()** (or its companion **pd.notnull()**)

```python
reviews[pd.isnull(reviews.country)]
```

Replacing missing values is a common operation.  
Pandas provides a nice method for this problem: **fillna()** - this provides a few different strategies for mitigating such data 

```python
#we can replace each NaN with an "Unknown":
reviews.region_2.fillna('Unknown')
```

alternatively we can fill each missing value with the first non-null value that appears sometime after the given record in the database.    
this is known as the **backfill** strategy  
the **forwardfill** strategy fills the null value with the last non-null value

we may have a non-null value that we want to replace - the **replace()** method is used 

```python
#suppose @kerinokeefe changed her handle to @kerino
reviews.taster_twitter_handle.replace('@kerinokeefe','@kerino')
```

the _replace()_ function is worth noting here because it's handy for replacing missing data which is given some kind of sentinel value in the dataset like 'Unknown', 'Undisclosed', 'Invalid' etc...
