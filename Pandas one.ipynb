## Indexing

i) iloc ii) loc

#### i) Index-based selection

indexing works in one of two paradigm, the first is index-based location, selecting data based on its  
numerical position in the data.  
**iloc** uses this

```python
reviews.iloc[0] #selects the first row of a dataframe 
```

**both iloc and loc are row first, column second**

```python
reviews.iloc[:, 0] #gets the first column from a df
# : means everything 
reviews.iloc[:3, 0] # selects the first 3 rows in the first column 

#it's also possible to pass a list 
reviews.iloc[[0,1,2], 0]
```

- _*like indexing lists, negative numbers can be used to select the last elements_

#### ii) Label based selection 
the data index value not its position is what matters  
**loc** is used

```python
review.loc['country',0] #selects the first country in the df
```

iloc treats the whole df like a big matrix (a list of lists), that we index inot by position  
loc uses the info in indicies to do the work, this makes it easier to use eg:

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']] #this is far easier to do using loc than iloc
```

##### choosing between iloc and loc
-iloc uses the stdlib scheme, where the first element is included and the last term is excluded, 0:10 will select 0,1,2...,9  
-loc indexes inclusively, 0:10 will select 0,1,2...,10

## Conditional Selection



```python
reviews.country=='Italy' #checks if each line in the country column is Italian or not 
```

this operation produces a series of True/False booleans based on the country of each record   
this result can then be used inside _loc_ to select the relevant data 

```python
reviews.loc[reviews.country=='Italy'] #produces a df of wines from Italy only
```

if you want to prod a df of good quality wines(with points >= 90) from Italy you can use **&**

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)] #creates a df of wines from Italy with a rating of 90 or greater 
```

- **_&_** _does a row by row operation and returns a value whereas_ **_and_** _returns a boolean. Same as_ **_or_** _and_ **_|_**

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

i) _isin_ - selects data where the a value 'is in' a list of values

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])] #selects wines from Italy or France
```

ii) _isnull (and notnull)_ - highlights values which are empty(or not)

```python
reviews.loc[reviews.price.notnull()] #filters out wines lacking a pricetag in the dataset
```
