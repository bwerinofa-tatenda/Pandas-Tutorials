# Renaming and combining

a lot of times we get data with column names, index names or other naming conventions that we dont want to use

## Renaming
**rename()** will let you change index names and/or column names.  
e.g to change the `points` column in the dataset to _score_ we would:

```python
reviews.rename(columns={'points': 'score'})
```

_rename()_ lets you rename index or column values by specifying a index or column keyword parameter, respectively.  
it support a variety of input formats, but usually a Python dictionary is the most convenient

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

remaning columns is more common, resetting index values is rare and for that **set_index()** is more convenient.  
both the row and the column index can have their own _name_ attribute. The complimentary **rename_axis()** method can be used to change these names

```python
reviews.rename_axis('wines', axis='rows').
rename_axis('feilds', axis='columns')
```

## Combining
when performing operations on a dataset we sometimes need to combine different DataFrames and/series in non-trivial ways.  
Pandas has 3 core methods for doing this. In order of increasing complexity these are:  
- **concat()**
- **join()**
- **merge()**  
most of what merge does can also be done by more simply with `join()`  

this is useful when we have datasets in different DataFrame or Series objects but having the same feilds(columns)

one example is the YouTube Videos dataset, which splits the data up based on country of origin.  
if we want to study multiple countries simultaneously, we can use `concat()` to smush them together

```python
pd.concat([canadian_youtube, british_youtube])
```

`join()` lets you combine different DataFrame objects which have a common index.    
e.g to pull down videos that happened to be trending on the same day in both Canada and the UK, we could use:

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```

`lsuffix` and `rsuffix` parameters are necessary here because the data has the same column names in both British and Canadian datasets. if this wasnt true(suppose we had renamed them beforehand) we wouldn't need them
