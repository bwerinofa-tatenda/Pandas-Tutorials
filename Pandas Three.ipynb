## Grouping and Sorting 

Maps allow us to transform data in a DataFrame or Series one value at a time for an entire column  
but we often want to group our data and do something specific to the group the data is in  

this is done by the **_groupby()_** operation

1. #### Groupwise analysis

```python
#we can replicate the value_counts() function by:
reviews.groupby('points').points.count()
```

_groupby_ in the above example created a group of reviews which allotted the same point values given wines. then for each of these groups, we grabbed the points() column and counted how many times it appeared. *value_counts*() is a shortcut to this groupby operation.

we can also use summary functions with groupby()

```python
reviews.groupby('points').price.min()
```

think of each group we generate as being a slice of our DataFrame containing only data with values that match.  
This DataFrame is accessible to us directly using the **_apply()_** method, and we can then manipulate the data in any way we see fit. 

```python
#e.g heres one way of selecting the name of the first wine reviewed from each winery in the dataset:
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])

#you can also group by more than one column:
reviews.groupby(['country', 'province']).apply(lambda df: 
                                               df.loc[df.points.idxmax()])
#this line of code picks out the best wine by country and province
```

_agg()_ is a groupby() method that lets you run a bunch of different functions on your DataFrame simultaneously.  
e.g we can generate a simple simple statistical summary of the dataset:

```python
reviews.groupby(['country']).price.agg([len,min,max])
```

## Multi-index
all the examples weve done so far the Dataframe or Series objects had a single-label index. 
using _groupby()_ sometimes results in a mutli-index 

```python
#an example of an operation that gives multiple indices
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
countries_reviewed.reset_index()
```

## Sorting
grouping returns data in index order not in value order - outputting a result of a _groupby()_ the order of the rows is dependend on the values in the index not in the data
to get the data in the order we want we can sort it ourselves.
**sort_values()** is handy for this 

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by=len)
#sort_values() defaults to an ascend sort, where the lowest values go first

#when we want a descending sort:
countries_reviewed.sort_values(by=len, ascending=False)
```

to sort by index values, use the method **sort_index()** 

```python
#this method uses the same arguments and default order
countries_reviewed.sort_values(by=['country', 'len'])
```
